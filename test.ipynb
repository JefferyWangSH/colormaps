```python
%reset -f

import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget

from colormaps import mma_rainbow

x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) + np.cos(Y)

fig, ax = plt.subplots(1, 1, figsize=(3,3))
im = ax.contourf(X, Y, Z, levels=10, cmap=mma_rainbow)
fig.colorbar(im, shrink=.7)
ax.set_box_aspect(1)
fig.tight_layout()
```
